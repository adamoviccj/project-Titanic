```python
def obradi(kolone):
    Age=kolone.iloc[0]
    Pclass=int(kolone.iloc[1])

    if pd.isnull(Age):
        print(f"{Age}, {Pclass}, {type(Age)}")
        if Pclass == 1:
            return 27
        elif Pclass == 2:
            return 29
        elif Pclass == 3:
            return 25
    else:
        return Age
```

```python
df['Age']=df[['Age', 'Pclass']].apply(obradi, axis=1)
```
