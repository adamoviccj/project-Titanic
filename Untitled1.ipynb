```python
# ukoliko utvrdimo da negde nedostaje podatak o godinama putnika, prazno mesto cemo popuniti prosecnim godinama putnika
# u zavisnosti od klase u kojoj je putovao
def obradi(cols):
    Age = cols.iloc[0]
    Pclass = int(cols.iloc[1])

    if pd.isnull(Age):
        print(f"{Age}, {Pclass}, {type(Age)}")
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        elif Pclass == 3:
            return 25
    else:
        return Age
```

```python
def plot_confusion_matrix(y_true, y_pred, class_names):
    """
    Funkcija za crtanje grafika matrice konfuzije koristeÄ‡i samo matplotlib.
    
    :param y_true: Lista stvarnih vrednosti
    :param y_pred: Lista predikovanih vrednosti
    :param class_names: Lista naziva klasa
    """
    # IzraÄunavanje matrice konfuzije
    cm = confusion_matrix(y_true, y_pred)
    
    # Kreiranje grafika
    fig, ax = plt.subplots(figsize=(10, 7))
    
    # Kreiranje imshow grafika
    cax = ax.matshow(cm, cmap=plt.cm.Blues)
    fig.colorbar(cax)
    
    # Postavljanje osa
    ax.set_xticks(np.arange(len(class_names)))
    ax.set_yticks(np.arange(len(class_names)))
    ax.set_xticklabels(class_names)
    ax.set_yticklabels(class_names)
    
    # ObeleÅ¾avanje osovina
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    
    # Dodavanje vrednosti u Ä‡elije matrice
    for i in range(len(class_names)):
        for j in range(len(class_names)):
            ax.text(j, i, cm[i, j], ha='center', va='center', color='black')
    
    # Naslov
    plt.title('Confusion Matrix')
    plt.show()
```

```python
def print_statistics(y_true, y_pred, model_name):
    print(f"Statistics for model {model_name}")
    
    print(f"Accuracy score: {accuracy_score(y_true, y_pred)}")
    
    print(f"Confusion matrix")
    plot_confusion_matrix(y_true, y_pred, [0, 1])

    print(f"Precision score: {precision_score(y_true, y_pred)}")

    print(f"Sensitivity: {recall_score(y_true, y_pred)}")

    print(f"F1 Score: {f1_score(y_true, y_pred)}")
```

```python
def plot_metrics(history):
    loss=history.history['loss']
    val_loss=history.history['val_loss']
    accuracy=history.history['accuracy']
    val_accuracy=history.history['val_accuracy']

    plt.plot(loss)
    plt.plot(val_loss)
    plt.title("Loss")
    plt.legend(labels=['loss','val_loss'])
    plt.show()


    plt.plot(accuracy)
    plt.plot(val_accuracy)
    plt.title("Accuracy")
    plt.legend(labels=['accuracy','val_accuracy'])
    plt.show()




```
